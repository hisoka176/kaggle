```python
import numpy as np
import theano.tensor as T
from theano import shared
from theano import In
from theano import function
from theano import *
```

```python
rng = numpy.random
N = 400  # train sample size
feats = 784 # number of input variables 

# generate dataset :D=(input_values,target_class)
D = (rng.randn(N,feats),rng.randint(size=N,low=0,high=2))
training_steps = 1000

x = T.dmatrix('x')
y = T.dvector('y')

# initial the weight vector w randomly
#
# this and follow  bias variable b
# is shared so they keep their value 
# between training iterations (updates)

w = theano.shared(rng.randn(feats),name='w')
#initial bias term
b = theano.shared(0.,name='b')

print("inital the model:")

print(w.get_value())
print(b.get_value())

# construct theano expression graph
p_1 = 1/(1+T.exp(-T.dot(x,w)-b)) # probabilty tartget =1 
prediction = p_1 > 0.5  # the prediction threshold


xent = -y*T.log(p_1)-(1-y)*T.log(1-p_1) # cross-entropy losss function 
cost = xent.mean() + 0.01*(w**2).sum() # the cost to mimize 

gw,gb = T.grad(cost,[w,b])  # compute the gradient of cost 
                            # 

#compile
logistic = theano.function(
    inputs=[x,y],
    outputs = [w,b],
    updates =[(w,w-0.1*gw),(b,b-0.1*gb)])
predict = theano.function(inputs=[x],outputs=prediction)

#train 
for i in range(training_steps):
    pre,err = logistic(D[0],D[1])

print("last model ")
print(w.get_value())
print(b.get_value())

print("target values for D:")
print(D[1])

print("predict values on D:")
pre = predict(D[0])
print(pre)
```
